4. Scrape the details of State-wise GDP of India from statisticstime.com.
Url = http://statisticstimes.com/

You have to find following details:

A) Rank B) State C) GSDP at current price (19-20) D) GSDP at current price (18-19) E) Share(18-19) F) GDP($ billion)

```python
driver=webdriver.Edge(r"C:/Users/91945/Downloads/edgedriver_win64/msedgedriver.exe")
url="http://statisticstimes.com/economy/india/indian-states-gdp.php"
driver.get(url)
time.sleep(5)
Rank=[]
State =[]
GDP=[]
GSDP_Current=[]
GSDP_Previous=[]
Share=[]

#scraping the Rank 
r=driver.find_elements_by_xpath("//td[@class='data1']")
for i in r:
    if i.text is None :
        Rank.append("--") 
    else:
        Rank.append(i.text)
St=driver.find_elements_by_xpath("//td[@class='name']")
for i in St:
    if i.text is None :
        State.append("--") 
    else:
        State.append(i.text)
gdp=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[6]")
for i in gdp:
    if i.text is None :
        GDP.append("--") 
    else:
        GDP.append(i.text)
#scraping the Share 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[5]")
for i in shr:
    if i.text is None :
        Share.append("--") 
    else:
        Share.append(i.text)
#scraping the GSDP_Current 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[4]")
for i in shr:
    if i.text is None :
        GSDP_Current.append("--") 
    else:
        GSDP_Current.append(i.text)
#scraping the GSDP_Previous 
shr=driver.find_elements_by_xpath("//*[@id='table_id']/tbody/tr/td[8]")
for i in shr:
    if i.text is None :
        GSDP_Previous.append("--") 
    else:
        GSDP_Previous.append(i.text)
#Creating Dataframe
State_GDP=pd.DataFrame([])
State_GDP['Rank']=Rank[:33]
State_GDP['State']=State[:33]
State_GDP['Share In GDP']=Share[:33]
State_GDP['GDP of State']=GDP[:33]
State_GDP['GSDP_Current']=GSDP_Current[:33]
State_GDP['GSDP_Previous']=GSDP_Previous[:33]
State_GDP
```

5. Scrape the details of trending repositories on Github.com.
Url = https://github.com/

You have to find the following details:

A) Repository title B) Repository description C) Contributors count D) Language used

```python
driver=webdriver.Edge(r"C:/Users/91945/Downloads/edgedriver_win64/msedgedriver.exe")
url =  'https://github.com/trending'
driver.get(url)



Repo_title = []
Repo_desc = []
count = []
repourl = []
Repo_title_tag = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for r in Repo_title_tag:
    try:
        repourl.append(r.get_attribute("href"))
    except NoSuchElementException:
        repourl.append("-")
Repo_title_tag = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for r in Repo_title_tag:
    try:
        Repo_title.append(r.text)
    except NoSuchElementException:
        Repo_title.append("-")

for i in repourl:
    driver.get(i)
    time.sleep(5)
    try:   
        Repo_desc_tag = driver.find_element_by_xpath("//p[@class='f4 mt-3']")
        Repo_desc.append(Repo_desc_tag.text)
    except NoSuchElementException:
        Repo_desc.append('-')
    try:
        count_tag = driver.find_element_by_xpath("//a[contains(@href,'contributors')]/span")
        count.append(count_tag.text)
    except NoSuchElementException:
        count.append('-')
lng = []
for i in repourl:
    driver.get(i)
    time.sleep(5)
    try:
        lng_tag = driver.find_element_by_xpath("//a[contains(@href,'search?')]/span[1]")
        lng.append(lng_tag.text)
    except NoSuchElementException:
        lng.append('-')
Git = {'Repository title': Repo_title,'Repository description':Repo_desc,'Contributors count':count,'Language used':lng}
GitDF = pd.DataFrame(Git)
GitDF
```

6. Scrape the details of top 100 songs on billboard.com.
Url = https://www.billboard.com/

You have to find the following details:

A) Song name B) Artist name C) Last week rank D) Peak rank E) Weeks on board

```python
driver=webdriver.Edge(r"C:/Users/91945/Downloads/edgedriver_win64/msedgedriver.exe")
url =  "https://www.billboard.com/charts/hot-100"
driver.get(url)
Song_Name =[]
Singer=[]
rank=[]
Last_Week=[]
Weeks_on_board=[]

#scraping the Rank 
rb=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in rb:
    if i.text is None :
        rank.append("--") 
    else:
        rank.append(i.text)

#scraping the Song_Name 
son=driver.find_elements_by_xpath("//span[@class='chart-element__information__song text--truncate color--primary']")
for i in son:
    if i.text is None :
        Song_Name.append("--") 
    else:
        Song_Name.append(i.text)
#scraping the Singer 
sin=driver.find_elements_by_xpath("//span[@class='chart-element__information__artist text--truncate color--secondary']")
for i in sin:
    if i.text is None :
        Singer.append("--") 
    else:
        Singer.append(i.text)
#scraping the Last_Week Rank 
lwr=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in lwr:
    if i.text is None :
        Last_Week.append("--") 
    else:
        Last_Week.append(i.text)
#scraping the Weeks_on_board 
wob=driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in wob:
    if i.text is None :
        Weeks_on_board.append("--") 
    else:
        Weeks_on_board.append(i.text)
BillBOARD_HOT100=pd.DataFrame([])
BillBOARD_HOT100['Peak_rank']=rank
BillBOARD_HOT100['Song_Name']=Song_Name
BillBOARD_HOT100['Singer / Crew']=Singer
BillBOARD_HOT100['Last_Week_Rank']=Last_Week
BillBOARD_HOT100['Weeks_on_board']=Weeks_on_board
BillBOARD_HOT100
```
